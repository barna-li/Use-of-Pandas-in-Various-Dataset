## 10. who is the third most active person in this dataset find out those in pandas 

```python
#The person with first and second most total steps completed belongs to one id so segregating it 
#and looking into other unique id thereon we get our third most active person id.
Third_active= df.nlargest(n=8,columns=['TotalSteps'])
Third_active
```
