## 9 . Average weekly response time for each agent 

```python
b[['hh','mm','ss']]=b["Average Response Time"].str.split(":",expand=True)

b['hh']=b['hh'].astype(int)
b['mm']=b['mm'].astype(int)
b['ss']=b['ss'].astype(int)

b["response_in_sec"]=(b['hh']*3600) + (b['mm']*60) + (b['ss'])
b.head()
```

```python
#Response Time in seconds
group = b.groupby(["Agent Name"]).sum()/5
res = group["response_in_sec"]
res.sort_values(ascending = False)
```

## 10 . Average weekely resolution time for each agents 

```python
b[['hh','mm','ss']]=b["Average Resolution Time"].str.split(":",expand=True)

b['hh']=b['hh'].astype(int)
b['mm']=b['mm'].astype(int)
b['ss']=b['ss'].astype(int)

b["resolution_in_sec"]=(b['hh']*3600) + (b['mm']*60) + (b['ss'])
b.head()
```

```python
#Resolution time in seconds
group = b.groupby(["Agent Name"]).sum()/5
resolve = group["resolution_in_sec"]
resolve.sort_values(ascending = False)
```
